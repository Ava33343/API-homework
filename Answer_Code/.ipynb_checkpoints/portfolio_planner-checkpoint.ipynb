# Monte Carlo Simulation

In this step, you will run Monte Carlo Simulations for your portfolio to model portfolio performance at different retirement ages. 

Complete the following steps:
1. Calculate the daily returns for the SPY and AGG closing prices.
2. Calculate volatility for both the SPY and AGG closing prices.
3. Find the last day's closing price for both stocks and save those as variables.
4. Run a Monte Carlo Simulation of at least 500 iterations and generate at least 30 years of closing prices

### HINTS:
There are 252 trading days per year, so the number of records to generate for each Monte Carlo run will be 252 days * 30 years

```python
# Calculate the daily roi for the stocks
daily_returns = historical_df.pct_change()
daily_returns.head()
```

```python
# Use the `mean` function to calculate the mean of daily returns for `SPY` and `AGG`, respectively
avg_daily_return_SPY = daily_returns.mean()['SPY']['close']
avg_daily_return_AGG = daily_returns.mean()['AGG']['close']
avg_daily_return_AGG
```

```python
# Use the `std` function to calculate the standard deviation of daily returns for `JNJ` and `MU`, respectively
std_dev_daily_return_SPY = daily_returns.std()['SPY']['close']
std_dev_daily_return_AGG = daily_returns.std()['AGG']['close']
```

```python
std_dev_daily_return_SPY
```

```python
std_dev_daily_return_AGG
```

```python
# Calculate volatility, annualized standard deviation, based on 252 trading days in a year
volatility_SPY = std_dev_daily_return_SPY * np.sqrt(252)
volatility_AGG = std_dev_daily_return_AGG * np.sqrt(252)
```

```python
volatility_SPY
```

```python
volatility_AGG
```

```python
# Save the last day's closing price
SPY_last_price = historical_df['SPY']['close'][-1]
AGG_last_price = historical_df['AGG']['close'][-1]
```

```python
SPY_last_price
```

```python
AGG_last_price
```

```python
# Setup the Monte Carlo Parameters
number_simulations = 500
number_records = 252 * 30

# Initialize empty DataFrame to hold simulated prices for SPY & AGG 
# Initialize empty DataFrame for the cumulative returns of the simulated 60/40 portfolio
monte_carlo_prices = pd.DataFrame()
monte_carlo_portfolio = pd.DataFrame()
```

```python
# Run the Monte Carlo Simulation
for x in range(number_simulations):
    
    # Initialize simulated prices
    simulated_SPY_prices = [SPY_last_price]
    simulated_AGG_prices = [AGG_last_price]
    
    # Simulate the returns for the next 30 years
    for i in range(number_records):
        
        simulated_SPY_price = simulated_SPY_prices[-1] * (1 + np.random.normal(avg_daily_return_SPY, std_dev_daily_return_SPY))
        simulated_AGG_price = simulated_AGG_prices[-1] * (1 + np.random.normal(avg_daily_return_AGG, std_dev_daily_return_AGG))
        
        # Append the simulated price to the list
        simulated_SPY_prices.append(simulated_SPY_price)
        simulated_AGG_prices.append(simulated_AGG_price)
    
    # Append a simulated prices of each simulation to DataFrame
    monte_carlo_prices["SPY prices"] = pd.Series(simulated_SPY_prices)
    monte_carlo_prices["AGG prices"] = pd.Series(simulated_AGG_prices)
    
    # Calculate the daily returns of simulated prices
    simulated_daily_returns = monte_carlo_prices.pct_change()
    
    # Set the portfolio weights (60% SPY; 40% AGG)
    weights = [0.60, 0.40]

    # Use the `dot` function with the weights to multiply weights with each column's simulated daily returns
    portfolio_daily_returns = simulated_daily_returns.dot(weights)
    
    # Calculate the normalized, cumulative return series
    monte_carlo_portfolio[x] = (1 + portfolio_daily_returns.fillna(0)).cumprod()

# Print records from the DataFrame
monte_carlo_portfolio.head()
```

### Given the current projected annual income from the Plaid analysis, will a 4% withdraw rate from the retirement portfolio meet or exceed that value at the 10th percentile?

Note: This is effectively saying that 90% of the expected returns will be greater than the return at the 10th percentile, so this can help measure the uncertainty about having enough funds at retirement

```python
# Based on Plaid analysis, the projected annual income is $6,054.80.
projected_yearly_income = 6_054.80
withdrawal = investment_pnl_10th_percentile * 0.04

if withdrawal > projected_yearly_income:
    print(f"At the end of 30 years, there is a 90% chance that the 4% withdrawal of ${withdrawal:,.2f} from the retirement portfolio" 
          f" is greater than the projected annual income of ${projected_yearly_income:,.2f}.")
else:
    print(f"A 4% withdrawal of ${withdrawal:,.2f} from the retirement portfolio" 
          f" does not meet the projected annual income of ${projected_yearly_income:,.2f} for 90% of the time.")
```

### A portfolio with more risk - 90% stocks (SPY) and 10% bonds (AGG)

```python
# Setup the Monte Carlo Parameters
number_simulations_early_retirement = 500
number_records_early_retirement = 252 * 5

# Initialize empty DataFrame to hold simulated prices for SPY & AGG 
# Initialize empty DataFrame for the cumulative returns of the simulated 60/40 portfolio
monte_carlo_prices_early_retirement = pd.DataFrame()
monte_carlo_portfolio_early_retirement = pd.DataFrame()
```

```python
# Run the Monte Carlo Simulation 
for n in range(number_simulations_early_retirement):
    
    # Initialize simulated prices
    simulated_SPY_prices_early_retirement = [SPY_last_price]
    simulated_AGG_prices_early_retirement = [AGG_last_price]
    
    # Simulate the returns for the next 30 years
    for y in range(number_records_early_retirement):
        
        simulated_SPY_price_early_retirement = simulated_SPY_prices_early_retirement[-1] * (1 + np.random.normal(avg_daily_return_SPY, std_dev_daily_return_SPY))
        simulated_AGG_price_early_retirement = simulated_AGG_prices_early_retirement[-1] * (1 + np.random.normal(avg_daily_return_AGG, std_dev_daily_return_AGG))
        
        # Append the simulated price to the list
        simulated_SPY_prices_early_retirement.append(simulated_SPY_price_early_retirement)
        simulated_AGG_prices_early_retirement.append(simulated_AGG_price_early_retirement)
    
    # Append a simulated prices of each simulation to DataFrame
    monte_carlo_prices_early_retirement["SPY prices"] = pd.Series(simulated_SPY_prices_early_retirement)
    monte_carlo_prices_early_retirement["AGG prices"] = pd.Series(simulated_AGG_prices_early_retirement)
    
    # Calculate the daily returns of simulated prices
    simulated_daily_returns_early_retirement = monte_carlo_prices_early_retirement.pct_change()
    
    # Set the portfolio weights (90% SPY; 10% AGG)
    weights_early_retirement = [0.90, 0.10]

    # Use the `dot` function with the weights to multiply weights with each column's simulated daily returns
    portfolio_daily_returns_early_retirement = simulated_daily_returns_early_retirement.dot(weights_early_retirement)
    
    # Calculate the normalized, cumulative return series
    monte_carlo_portfolio_early_retirement[n] = (1 + portfolio_daily_returns_early_retirement.fillna(0)).cumprod()

# Print records from the DataFrame
monte_carlo_portfolio_early_retirement.head()
```

```python
# Use the `plot` function to plot `500` simulations of the potential trajectories of the portfolio based on 252*5 trading days
plot_title = f"Figure 8. {x+1} Simulations of Daily Cumulative Portfolio Return Trajectories Over the Next 5 Years"
monte_carlo_portfolio_early_retirement.plot(legend=None, title=plot_title, figsize=(12,7))
```

```python
# Create an empty dataframe for medium and 90% confidence interval on daily cumulative returns over the next 5 years

monte_carlo_daily_briefing_early_retirement = pd.DataFrame()

for n in monte_carlo_portfolio_early_retirement:
    
    median_early_retirement = monte_carlo_portfolio_early_retirement.median(axis=1)
    confidence_interval_early_retirement_90_lower = monte_carlo_portfolio_early_retirement.quantile(0.95, axis=1)
    confidence_interval_early_retirement_90_upper = monte_carlo_portfolio_early_retirement.quantile(0.05, axis=1)

    # Append median and bounds for 90% confidence interval to the dataframe
    monte_carlo_daily_briefing_early_retirement["CI_90_Lower"] = pd.Series(confidence_interval_early_retirement_90_lower)
    monte_carlo_daily_briefing_early_retirement["Median"] = pd.Series(median)
    monte_carlo_daily_briefing_early_retirement["CI_90_Upper"] = pd.Series(confidence_interval_early_retirement_90_upper)

monte_carlo_daily_briefing_early_retirement.tail()
```

```python
# Use the `plot` function to plot `500` simulations of the potential trajectories of the portfolio median and two bounds based on 252*5 trading days
plot_title = f"Figure 9. {x+1} Median and 90% Confidence Interval Trajectories on Simulations of Cumulative Daily Portfolio Return Over the Next 30 Years"
monte_carlo_daily_briefing_early_retirement.plot(legend=None, title=plot_title, figsize=(12,7))
```
